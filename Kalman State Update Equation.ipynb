```python
import numpy as np
import matplotlib.pyplot as plt
```

## Syestem State Update Equation
$$ \hat{x}_\mathrm{n,n}= \hat{x}_\mathrm{n,n-1} + \alpha_\mathrm{n}(z_\mathrm{n} - \hat{x}_\mathrm{n,n-1}) $$
### Gold Bar Example
Initial guess:
$$  \hat{x}_\mathrm{0,0}= 1000g $$
First Measurment:
$$ z_\mathrm{1} = 996

```python
def state_update_equation(x_predicted,n,z_measurment):
    return (x_predicted + ((1/n)*(z_measurment-x_predicted)))
```

```python
# Iteration 1
x_initial_guess = 1000
z1 = 996
n = 1
x_n = state_update_equation(x_initial_guess,n,z1)
print('1st itaration xn,n = ',x_n, 'g')
# Iteration 2
z2 = 994
x_n1 = state_update_equation(x_n,2,z2)
print('2nd itaration xn,n = ',x_n1, 'g')
# Iteration 3
z3 = 1021
x_n2 = state_update_equation(x_n1,3,z3)
print('3rd itaration xn,n = ',x_n2, 'g')
# Iteration 4
z4 = 1000
x_n3 = state_update_equation(x_n2,4,z4)
print('4th itaration xn,n = ',x_n3, 'g')
# Iteration 5
z5 = 1002
x_n4 = state_update_equation(x_n3,5,z5)
print('5th itaration xn,n = ',x_n4, 'g')
# Iteration 6
z6 = 1010
x_n5 = state_update_equation(x_n4,6,z6)
print('6th itaration xn,n = ',x_n5, 'g')
# Iteration 7
z6 = 983
x_n6 = state_update_equation(x_n5,7,z6)
print('7th itaration xn,n = ',x_n6, 'g')
# Iteration 8
z7 = 971
x_n7 = state_update_equation(x_n6,8,z7)
print('8th itaration xn,n = ',x_n7, 'g')
# Iteration 9
z8 = 993
x_n8 = state_update_equation(x_n7,9,z8)
print('9th itaration xn,n = ',x_n8, 'g')
# Iteration 10
z9 = 1023
x_n9 = state_update_equation(x_n8,10,z9)
print('10th itaration xn,n = ',x_n9, 'g')
```

```python
# Simplify the code using for loop:
x_initial_guess = 1000
measurements = [996,994,1021,1000,1002,1010,983,971,993,1023]
actual = 1000 * np.ones(20)
x_n = x_initial_guess
x_n_val = []
for i,z in enumerate(measurements,1):
    x_n = state_update_equation(x_n,i,z)
    x_n_val.append(x_n)
    print(f'{i}th iteration Xn,n= {round(x_n,3)} g')

    
plt.figure()
plt.plot(x_n_val, marker='o')
plt.plot(measurements, marker='o')
plt.plot(actual, marker='o')
plt.legend(['Estimates','Measurements','True Value'])
plt.xlabel('Iterations')
plt.ylabel('Weight (g)')
plt.title('Kalman Estimation Algorithm')
plt.grid()
```
