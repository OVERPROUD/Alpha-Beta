# Alpha-Beta Filter
State Extrapolation Equation 
$$ x_\mathrm{n+1}= x_\mathrm{n} + \Delta{t}\dot{x}_\mathrm{n} $$
$$ \dot{x}_\mathrm{n+1}= \dot{x}_\mathrm{n} $$
State Update Equation for Position
$$ \hat{x}_\mathrm{n,n}= \hat{x}_\mathrm{n,n-1} + \alpha(z_\mathrm{n} - \hat{x}_\mathrm{n,n-1}) $$
State Update Equation for Velocity
$$ \dot{x}_\mathrm{n,n}= \dot{x}_\mathrm{n,n-1} + \beta(\frac{z_\mathrm{n} - \hat{x}_\mathrm{n,n-1}}{\Delta{t}}) $$ 
## TRACKING THE CONSTANT VELOCITY AIRCRAFT IN ONE DIMENSION
The radar track-to-track period is
$$ \Delta{t}=5s $$
The Estimated Range of the UAV is 
$$  \hat{x}_\mathrm{0,0}= 30000m $$
The Estimated UAV Velocity is 
$$ \dot{x}_\mathrm{0,0}= 40m/s $$

```python
def state_extrapolation_equation(xn,dt,xdotn):
    return (xn + dt*xdotn)
def state_update_equation_for_position(x_predicted,alpha,z_measurment):
    return (x_predicted + (alpha*(z_measurment-x_predicted)))
def state_update_equation_for_velocity(v_predicted,beta,z,x_predicted,dt):
    return (v_predicted+(beta*((z-x_predicted)/dt)))
```

```python
x0 = 30000
dt = 5
v0 = 40
alpha = 0.2
beta = 0.1


x_predicted = state_extrapolation_equation(x0,dt,40)
v_predicted = v0


# First Iteration
z1 = 30171
x1 = state_update_equation_for_position(x_predicted,alpha,z1)
v1 = state_update_equation_for_velocity(v_predicted,beta,z1,x_predicted,dt)
xp1 = state_extrapolation_equation(x1,dt,v1)
print('1st itaration\n The current estimate position is ',x1, 'm\n',
      'The current estimate velocity is ',v1,'m/s\n','The next state estimate position is ',xp1,'m\n','The next state estimate velocity is ',v1,'m/s\n')

# Second Iteration
z2 = 30353
x2 = state_update_equation_for_position(xp1,alpha,z2)
v2 = state_update_equation_for_velocity(v1,beta,z2,xp1,dt)
xp2 = state_extrapolation_equation(x2,dt,v2)
print('2nd itaration\n The current estimate position is ',x2, 'm\n',
      'The current estimate velocity is ',v2,'m/s\n','The next state estimate position is ',xp2,'m\n','The next state estimate velocity is ',v2,'m/s\n')

# Third Iteration
z3 = 30756
x3 = state_update_equation_for_position(xp2,alpha,z3)
v3 = state_update_equation_for_velocity(v2,beta,z3,xp2,dt)
xp3 = state_extrapolation_equation(x3,dt,v3)
print('3rd itaration\n The current estimate position is ',x3, 'm\n',
      'The current estimate velocity is ',v3,'m/s\n','The next state estimate position is ',xp3,'m\n','The next state estimate velocity is ',v3,'m/s\n')

# Fourth Iteration
z4 = 30799
x4 = state_update_equation_for_position(xp3,alpha,z4)
v4 = state_update_equation_for_velocity(v3,beta,z4,xp3,dt)
xp4 = state_extrapolation_equation(x4,dt,v4)
print('4th itaration\n The current estimate position is ',x4, 'm\n',
      'The current estimate velocity is ',v4,'m/s\n','The next state estimate position is ',xp4,'m\n','The next state estimate velocity is ',v4,'m/s\n')

# Fifth Iteration
z5 = 31018
x5 = state_update_equation_for_position(xp4,alpha,z5)
v5 = state_update_equation_for_velocity(v4,beta,z5,xp4,dt)
xp5 = state_extrapolation_equation(x5,dt,v5)
print('5th itaration\n The current estimate position is ',x5, 'm\n',
      'The current estimate velocity is ',v5,'m/s\n','The next state estimate position is ',xp5,'m\n','The next state estimate velocity is ',v5,'m/s\n')

# Sixth Iteration
z6 = 31278
x6 = state_update_equation_for_position(xp5,alpha,z6)
v6 = state_update_equation_for_velocity(v5,beta,z6,xp5,dt)
xp6 = state_extrapolation_equation(x6,dt,v6)
print('6th itaration\n The current estimate position is ',x6, 'm\n',
      'The current estimate velocity is ',v6,'m/s\n','The next state estimate position is ',xp6,'m\n','The next state estimate velocity is ',v6,'m/s\n')

# Seventh Iteration
z7 = 31276
x7 = state_update_equation_for_position(xp6,alpha,z7)
v7 = state_update_equation_for_velocity(v6,beta,z7,xp6,dt)
xp7 = state_extrapolation_equation(x7,dt,v7)
print('7th itaration\n The current estimate position is ',x7, 'm\n',
      'The current estimate velocity is ',v7,'m/s\n','The next state estimate position is ',xp7,'m\n','The next state estimate velocity is ',v7,'m/s\n')

# Eighth Iteration
z8 = 31379
x8 = state_update_equation_for_position(xp7,alpha,z8)
v8 = state_update_equation_for_velocity(v7,beta,z8,xp7,dt)
xp8 = state_extrapolation_equation(x8,dt,v8)
print('8th itaration\n The current estimate position is ',x8, 'm\n',
      'The current estimate velocity is ',v8,'m/s\n','The next state estimate position is ',xp8,'m\n','The next state estimate velocity is ',v8,'m/s\n')

# Ninth Iteration
z9 = 31748
x9 = state_update_equation_for_position(xp8,alpha,z9)
v9 = state_update_equation_for_velocity(v8,beta,z9,xp8,dt)
xp9 = state_extrapolation_equation(x9,dt,v9)
print('9th itaration\n The current estimate position is ',x9, 'm\n',
      'The current estimate velocity is ',v9,'m/s\n','The next state estimate position is ',xp9,'m\n','The next state estimate velocity is ',v9,'m/s\n')

# Ninth Iteration
z10 = 32175
x10 = state_update_equation_for_position(xp9,alpha,z10)
v10 = state_update_equation_for_velocity(v9,beta,z10,xp9,dt)
xp10 = state_extrapolation_equation(x10,dt,v10)
print('10th itaration\n The current estimate position is ',x10, 'm\n',
      'The current estimate velocity is ',v10,'m/s\n','The next state estimate position is ',xp10,'m\n','The next state estimate velocity is ',v10,'m/s\n')
```

```python
x0 = 30000
dt = 5
v0 = 40
alpha = 0.2
beta = 0.1

z_measurment = [30171,30353,30756,30799,31018,31278,31276,31379,31748,32175]
xn = x0
vn = 40
xp_array = []
vn_array = []
xn_array = []
for i,z in enumerate(z_measurment,1):
    xp = state_extrapolation_equation(xn,dt,vn)
    xp_array.append(xp)
    xn = state_update_equation_for_position(xp,alpha,z)
    xn_array.append(xn)
    vn = state_update_equation_for_velocity(vn,beta,z,xp,dt)
    vn_array.append(vn)
    print(f' The next state estimate position is {round(xp,3)}m\n\n{i}th iteration\n The current estimate position is {round(xn,3)} m\n The current estimate velocity is {round(vn,3)} m/s\n The next state estimate velocity is {round(vn,3)}m/s')

xp = state_extrapolation_equation(xn,dt,vn)
print(f' The next state estimate position is {round(xp,3)}m')
t = np.arange(0,(5*10),5)
true_value = 30200 + t*40
plt.figure()
plt.plot(t,true_value,marker='o')
plt.plot(t,xn_array,marker='o')
plt.plot(t,z_measurment,marker='o')
plt.plot(t,xp_array,marker='o')
plt.legend(['True Value','Estimate','Measurment','Prediction'])
plt.xlabel('Time (s)')
plt.ylabel('Range (m)')
plt.title('Alpha-Beta Algorithm')
plt.grid()
```
